```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
poke = pd.read_csv('data/Pokemon.csv')
```

```python
poke.head()
```

```python
poke.describe()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(poke, x='Type 1', hue = 'Legendary')
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(poke, x='Legendary', y = 'Total')
```

```python
type_1 = pd.get_dummies(poke['Type 1']).astype(int)
print(type_1)
```

```python
poke_1 = pd.concat([poke, type_1], axis=1)  
```

```python
poke_1.drop('Type 1', axis=1, inplace=True)
poke_1.drop('Type 2', axis=1, inplace=True)
```
